```python
import pandas as pd
import re
```

```python
patt = re.compile('^(\d+)-(\d+) ([a-z]): (\S+)\n$')

def get_data():
    with open('input.txt') as f:
        for line in f:
            yield patt.findall(line)[0]
            
df = pd.DataFrame(get_data())
df.columns = ['low', 'high', 'char', 'password']
```

```python
from collections import Counter
```

```python
count = df.apply(lambda x: Counter(x.password)[x.char], axis=1)
```

```python
df.low = df.low.astype(int)
df.high = df.high.astype(int)
```

```python
# Part 1
df[(count >= df.low) & (count <= df.high)]
```

```python
# Part 2
```

```python
def compare(x):
    return (x.char == x.password[x.low - 1]) ^ (x.char == x.password[x.high - 1])   
```

```python
df[df.apply(compare, axis=1)]
```
